```python
import numpy as np
from os import listdir
from xml.etree import ElementTree
from numpy import zeros
from numpy import asarray
from numpy import expand_dims
from numpy import mean
from mrcnn.config import Config
from mrcnn.model import MaskRCNN
from mrcnn.utils import Dataset
from mrcnn.utils import compute_ap, compute_recall
from mrcnn.model import load_image_gt
from mrcnn.model import mold_image


CUSTOM_DATASET_DIR = '/content/'
class PredictionConfig(CustomConfig):
  NAME = "cell"
  #NUM_CLASSES = 1 + 1
  #LEARNING_RATE=0.0002
  #GPU_COUNT = 1
  #RPN_ANCHOR_RATIOS=[0.11, 0.25, 5.91, 12.58, 29.36]
  #IMAGES_PER_GPU = 1


def evaluate_model(dataset, model, cfg,i):
  APs = list();
  ARs = list();
  F1_scores = list();
  #print("F1")
  for image_id in dataset.image_ids:
    #print("F")
    if image_id%20==0:
      print(image_id)
    image, image_meta, gt_class_id, gt_bbox, gt_mask = load_image_gt(dataset, cfg, image_id, use_mini_mask=False)
    scaled_image = mold_image(image, cfg)
    sample = expand_dims(scaled_image, 0)
    yhat = model.detect(sample, verbose=0)
    r = yhat[0]
    #print("---------r--------")
    #print(r)
    AP, precisions, recalls, overlaps = compute_ap(gt_bbox, gt_class_id, gt_mask, r["rois"], r["class_ids"], r["scores"], r['masks'], iou_threshold=i)
    #print(precisions)
    AR, positive_ids = compute_recall(r["rois"], gt_bbox, iou=i)
    ARs.append(AR)
    F1_scores.append((2* (mean(precisions) * mean(recalls)))/(mean(precisions) + mean(recalls)))
    APs.append(AP)
	
  mAP = mean(APs)
  mAR = mean(ARs)
  F1=mean(F1_scores)
  return mAP, mAR, F1


test_set = CustomDataset(CUSTOM_DATASET_DIR)
#is_train=False
test_set.load_custom(CUSTOM_DATASET_DIR, 'test')
test_set.prepare()

#print(test_set.__dict__)

cfg = PredictionConfig()
model = MaskRCNN(mode='inference', model_dir='/content/drive/My Drive/Mask_Scoring_RCNN/', config=cfg)
model.keras_model.summary()
model.load_weights('/content/drive/MyDrive/mask_rcnn_cell_0100___icdar19_best results.h5', by_name=True)
model.keras_model.summary()
#print(model.summary())
l=[0.5,0.6,0.7,0.8,0.9]
l1=[]
for i in l:
  print("IOU")
  print(i)
  mAP, mAR, F1_score = evaluate_model(test_set, model, cfg,i)
  print("mAP: %.3f" % mAP)
  print("mAR: %.3f" % mAR)
  print("first way calculate f1-score: ", F1_score)


  F1_score_2 = (2 * mAP * mAR)/(mAP + mAR)
  print('second way calculate f1-score_2: ', F1_score_2)
  l1.append((mAP,mAR,F1_score,F1_score_2))
for j in l1:
  print(j)
```
